```python
import pandas as pd
```

```python
import numpy as np
```

```python
df= pd.read_csv('dulieuDTD.csv')
```

```python
cols= df.columns
cols
```

```python
Y= df[cols[-1]].values
Y= (Y!=0).astype(int)
Y
```

```python
X= []
for col in cols[:-1]:
    X.append(df[col].values.tolist())
```

```python
X= np.array(X)
X = np.concatenate((np.ones((1, X.shape[1])), X), axis = 0)
X= X.T #chuyen vi
X.shape
```

```python
X
```

```python
X.shape
```

```python
X_test= X[1000:, :]
Y_test= Y[1000:]

X= X[:1000,:]
Y= Y[:1000]
```

```python
X.shape, X_test.shape
```

```python
def sigmoid(s):
    return 1/(1 + np.exp(-s))

def logistic_sigmoid_regression(X, y, w_init, eta, tol = 1e-4, max_count = 10000):
    # X: (d,N) : N so dong, d so cot
    w = [w_init]   
    N = X.shape[1] # 1- dong
    d = X.shape[0] #0- cot
    count = 0
    check_w_after = 20 #kiem tra 20 dong de kiem tra co khac nhau khong
    while count < max_count:
        # mix data 
        mix_id = np.random.permutation(N)# permutation = trộn
        for i in mix_id:
            xi = X[:, i].reshape(d, 1)#chạy cột và lấy dòng i = lấy 1 quan sát
            yi = y[i]
            zi = sigmoid(np.dot(w[-1].T, xi)) #dot = nhân ma trận mxn * nxp= mxp
            w_new = w[-1] + eta*(yi - zi)*xi #w[-1] = giá trị trước đó 
            count += 1
            # stopping criteria
            if count%check_w_after == 0:                
                if np.linalg.norm(w_new - w[-check_w_after]) < tol: # k/c giữa 2 điểm đủ nhỏ 
                    return w 
            w.append(w_new)
    return w
#gọi hàm
eta = .05 
d = X.shape[1] #1 - cột
print(d)
w_init = np.random.randn(d, 1) #d - số lượng số cần lấy random, 1= giới hạn số( giá trị max)

w = logistic_sigmoid_regression(X.T, Y.T, w_init, eta) #Do định nghĩa ngược --> gọi hàm sử dụng X.T và Y.T
print(w[-1])
```

```python
w= w[-1]
w
```

```python
Y_hat= sigmoid(np.dot(w.T, X.T))
```

```python
Y_= Y_hat[0] >= 0.5
Y_
```

```python
sum(Y_ == Y)/len(Y) #sum(Các giá trị True)/ tổng số lượng
```

```python
Y_hat_test= sigmoid(np.dot(w.T, X_test.T))
Y_pred= Y_hat_test[0] >= 0.5
sum(Y_pred == Y_test)/len(Y_test) #sum(Các giá trị True)/ tổng số lượng
```
